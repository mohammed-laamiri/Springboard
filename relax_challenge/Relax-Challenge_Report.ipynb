Our task in this challenge was to “identify which factors predict future user adoption.” To that end, we
fit a Random Forest to 70% of our dataset, predicted on the remaining 30% ,
and extracted feature importances from the model. We also optimized for our model’s hyperparameters.

What we found was that we were able to achieve a high accuracy of about 88% on the test set.
When we looked at feature importances, here is what we found:

Our most important feature is **marketing_drip**. This feature was so important that when we removed it and
re-ran the algorithm again, our accuracy declined to 71%.

We also see that the creation_source_ORG_INVITE, creation_source_SIGNUP, and invited_by_user have almost the same importance. The other features are not nearly as important by comparison.
